```python
import rpy2.robjects as robjects
from rpy2.robjects import pandas2ri
from scipy.io import loadmat
import numpy as np
import ntpath
import scanpy.api as sc

def path_leaf(path):
    """Get path's tail from a filepath"""
    head, tail = ntpath.split(path)
    return tail or ntpath.basename(head) 

def savemat_as_df(matfile):
    prefix = path_leaf(matfile).rstrip('.mat')
    data = loadmat(matfile)
    print(data.keys())
    labels = data['true_labs']
    labels_unwrapped = np.array([x[0] for x in labels])
    np.save('../data/{}_matrix.npy'.format(prefix), data['in_X'])
    np.savetxt('../data/{}_matrix.txt'.format(prefix), data['in_X'])

    np.save('../data/{}_labels.npy'.format(prefix), labels_unwrapped)
    scanpy_df = sc.read('../data/{}_matrix.txt'.format(prefix))
    scanpy_df.obs['cell_groups'] = labels_unwrapped
    
    if 'Genes' in data.keys():
        genes = data['Genes']
        if len(genes) == 1:
            genes =  genes[0]
        assert len(genes) ==  data['in_X'].shape[1], 'genes_len: {} {}| data_shape: {}'.format(len(genes),  data['in_X'].shape)
        genes_unwrapped = np.array([x[0] if x else 'gene_{}'.format(index) for index, x in enumerate(genes) ])
        np.save('../data/{}_genes.npy'.format(prefix), genes_unwrapped)
        scanpy_df.var['genes'] = genes_unwrapped
        
        
    scanpy_df.write('../data/{}_anndata.h5ad'.format(prefix))
    
    
```

```python
savemat_as_df('../../SIMLR/MATLAB/data/Test_1_mECS.mat')
savemat_as_df('../../SIMLR/MATLAB/data/Test_2_Kolod.mat')
savemat_as_df('../../SIMLR/MATLAB/data/Test_3_Pollen.mat')


              
```
