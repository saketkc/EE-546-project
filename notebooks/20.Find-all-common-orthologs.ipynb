```python
%pylab inline
import pandas as pd
import os
import glob

species_tx2gene_map = ["btaurus",
                       "rnorvegicus",
                       "pabelii",
                       "mdomestica",
                       "mmulatta",
                       "ptroglodytes",
                       #"mmusculus",
                       "hsapiens",
                       "ggallus",
                       "oanatinus",
                       "ggorilla",
                       "ppaniscus"]
```

```python
ortholog_tsv_dir = '../ortholog_tsv/'
```

```python
master_df = pd.DataFrame()
for f in glob.glob('{}/*'.format(ortholog_tsv_dir)):
    second_species = f.split('/')[-1].replace('.tsv', '').split('_')[-1]
    df = pd.read_table(f)[['ensembl_gene_id', '{}_homolog_ensembl_gene'.format(second_species), '{}_homolog_orthology_type'.format(second_species)]]
    df = df[df['{}_homolog_orthology_type'.format(second_species)]=='ortholog_one2one']
    df = df.drop(columns=['{}_homolog_orthology_type'.format(second_species)])
    df = df.set_index('ensembl_gene_id')
    master_df = master_df.join(df, how='outer')
```

```python
master_df.to_csv('../ortholog_tsv/ortholog_one2one.tsv', sep='\t', index=True, header=True)
```

```python
master_df.dropna().to_csv('../ortholog_tsv/ortholog_one2one_not_nan.tsv', sep='\t', index=True, header=True)
```
